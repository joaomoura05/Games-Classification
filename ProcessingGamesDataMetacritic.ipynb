```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import json
import re
import datetime
```

```python
# Writing json contents in list for do treating

with open('gamesInfos.json', 'r') as arq:
    infos = json.load(arq)
     
print(len(infos), 'games were dowloaded') # 18598 games with your data saved 
```

```python
# Creating a dataframe

df = pd.DataFrame(infos)

df = df.rename(columns={'numbers_ratings': 'ratings_numbers'})

display(df)
```

## Exploratory analysis and data treating 

### Checking if have null data and treating them

```python
print('(rows, columns) ->', df.shape, '\n----------------------------')

print(df.isnull().sum())

df.loc[df['release'].isnull() == True]
```

```python
# Deleting the rows which show null values

df.dropna(subset=['release'], inplace=True)

print('(rows, columns) ->', df.shape, '\n----------------------------')

print(df.isnull().sum())

display(df.loc[df['release'].isnull() == True])

display(df)
```

### Formating column 'release'

```python
def convert_date(date_string):
    try:
        return datetime.datetime.strptime(date_string, '%b %d, %Y').strftime('%m/%d/%Y')
    except ValueError:
        # retorna o valor original caso ocorra o erro
        return date_string

df['release'] = df['release'].apply(convert_date)
```

```python
# Manipulating data using Regex

# Deleting words of columns 'ratings_numbers' and 'genre(s)'

regex_ratings_numbers = re.compile(' [A-Za-z]+')
df['ratings_numbers'] = df['ratings_numbers'].apply(lambda x: re.sub(regex_ratings_numbers , '', x))

regex_genre = re.compile('Genre[(]s[)]:.')
df['genre(s)'] = df['genre(s)'].apply(lambda x: re.sub(regex_genre, '', x))

display(df)
```

```python
# Replacing 'On GameFAQs' by '1-2'
# 'On GameFAQs' was a error collect on web scraping phase

df['players_number'] = df['players_number'].replace('On GameFAQs', '1-2 Player(s)')

display(df)
```

```python
# Changing data type

print(df.dtypes, '\n----------------------------')

df['metascore'] = df['metascore'].astype(int)
df['critic_reviews'] = df['critic_reviews'].astype(int)
df['user_score'] = df['user_score'].astype(float)
df['ratings_numbers'] = df['ratings_numbers'].astype(int)

print(df.dtypes)
```

```python
# Verifing quantities of unique values

print('Exist', df['title'].nunique(), 'games titles \n------------------------------------------')

print('Exist', df['plataform'].nunique(), 'games plataforms being:')
print(df['plataform'].unique(), '\n------------------------------------------')

print('Exist', df['developer'].nunique(), 'games developer being:')
print(df['developer'].unique(), '\n------------------------------------------')

print('Exist', df['genre(s)'].nunique(), 'games genre(s) being:')
print(df['genre(s)'].unique(), '\n------------------------------------------')

print('Exist', df['players_number'].nunique(), 'games players_number being:')
print(df['players_number'].unique())
```

```python
# Example
text = 'Action, Fighting, 3D'

# This function will be group by text and fix them
def genre_reduce(text):
    
    genre_list = re.findall(r"[0-9A-Z-a-z]+", text)

    genre_list = set(genre_list)
    
    text = ', '.join(genre_list)
    
    return text.replace(' ', '').replace(',', ', ')
genre_reduce(text)
```

```python
df['genre(s)'] = df['genre(s)'].apply(genre_reduce)

display(df)
```

```python
print('Exist', df['genre(s)'].nunique(), 'games genre(s) being:')
print(df['genre(s)'].unique(), '\n------------------------------------------')
```
